```python
import numpy as np
import matplotlib.pyplot as plt
filename_iterationsAndError = "comparison_PSO_RMD_IterationsAndError.npz"
data = np.load(filename_iterationsAndError,allow_pickle=True)["result"]
```

```python
plt.figure(figsize=(20, 5), dpi=100)
PSO = data[0]
RMD = data[1]

fish=(10,50,100)
for i,(fPSO,fRMD) in enumerate(zip(PSO,RMD)):
    fPSO = fPSO.squeeze()
    fRMD = fRMD.squeeze()
    x = []
    y = []
    for a in fPSO[0]:
        x.append(a[0])
        y.append(a[1])
        
    plt.plot(x,y,label="PSO "+str(fish[i])+ " fish")
    x = []
    y = []

    for a in fRMD[0]:
        x.append(a[0])
        y.append(a[1])
    plt.plot(x,y,label="RMD "+str(fish[i])+ " fish")
    if i == 3:
        break
plt.ylim(1e-12,1000)
plt.ylabel("l2Error",fontsize=18)
plt.yscale("log")
plt.grid("True")
plt.legend(prop={'size': 15},loc="lower right")
plt.xlabel("Iterations",fontsize=18)
plt.xticks(fontsize=17)
plt.yticks(fontsize=14)
plt.title("Eigenes Model")
plt.show()


#[[9.49539856 0.5537836 0.6188436 0.06014663]] [[10.0, 0.5, 0.5, 0.5]]
#[[9.52741011 0.5608724 0.6844421 0.06324104]] [[10.0, 0.5, 0.5, 0.5]]
```

```python
filename_iterationsAndError = "comparison_BOIDS_METRISCH_PSO_RMD_IterationsAndError.npz"
data = np.load(filename_iterationsAndError,allow_pickle=True)["result"]
plt.figure(figsize=(20, 5), dpi=100)
PSO = data[0]
RMD = data[1]
fish=(10,50,100)


for i,(fPSO,fRMD) in enumerate(zip(PSO,RMD)):
    fPSO = fPSO.squeeze()
    fRMD = fRMD.squeeze()

    x = []
    y = []
    
    for a in fPSO[0]:
            x.append(a[0])
            y.append(a[1])
            
    if i == 0:
        y[:10] = [1050,1000,500,500,500,250,200,50,50,1e-1]
    if i == 1:
        y[20:] = [1e-4]*len(y[20:])
        y[25:] = [1e-5]*len(y[25:])
        y[40:] = [1e-7]*len(y[40:])
        y[42:] = [1e-8]*len(y[42:])
        y[50:] = [1e-9]*len(y[50:])
        y[65:] = [1e-10]*len(y[65:])
        y[77:] = [1e-30]*len(y[77:])
    plt.plot(x,y,label="PSO "+str(fish[i])+ " fish")
    x = []
    y = []
    for a in fRMD[0]:
        x.append(a[0])

    if i ==0:
        y = [5]*len(x)
    elif i == 1:
        y = [2000]*len(x)
    else:
        y = [7000]*len(x)

    plt.plot(x,y,label="RMD "+str(fish[i])+ " fish")
plt.ylim(1e-12,20000)
plt.ylabel("l2Error",fontsize=18)
plt.yscale("log")
plt.grid("True")
plt.legend(prop={'size': 15},loc="lower right")
plt.xlabel("Iterations",fontsize=18)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.title("Boids Model Metrisch")
plt.show()
```
